```python
import pandas as pd

def get_dict(x):
  tmp_df2 = pd.DataFrame(x)
  tmp = {}
  for col in tmp_df2.columns:
    tmp[col] = list(tmp_df2[col].values)
  return tmp
def row_transformer(data):
    for key,value in data.items():
        #print (str(key)+'->'+str(value))
        if type(value) == type(dict()):
            row_transformer(value)
        elif type(value) == type(list()):
          tmp_dict = get_dict(value)
          print("tmp_dict:",tmp_dict)
          data[key] = tmp_dict
row_transformer(data)
```

## Normalize

```python
import pandas as pd
df = pd.json_normalize(data_blocks)
df
```

```python
def get_explode_columns():
  explode_column = []
  for column in df.columns:
    if type(df[column].iloc[0]) == str :
      continue
    else:
      if type(df[column].iloc[0]) == list:
        explode_column.append(column)
  return explode_column

explode_column = get_explode_columns()
explode_column
```

```python
def get_map_column(explode_column):
  map_column_list = []
  tmp_dict ={}
  flipped = {}
  for col in explode_column:
    tmp_dict[col] = len(df[col].iloc[0])

  # from dictionary using flip
  for key, value in tmp_dict.items():
      if value not in flipped:
          flipped[value] = [key]
      else:
          flipped[value].append(key)
  map_column_list = list(flipped.values())
  return tmp_dict, map_column_list

tmp_dict, map_column_list = get_map_column(explode_column)
tmp_dict, map_column_list
```

```python
def map_list_to_df(df, key_list,val_list):
  df_tmp = df[[key_list, val_list]].copy()
  for x in [key_list, val_list]:
    explode_columns = df[x].apply(lambda x: type(x)==list)
    df_output = pd.concat([df.loc[explode_columns].explode(column=[x]),
                   df.loc[~explode_columns]])
  first_entry = df_output[key_list].iloc[0]
  for i in range(0, len(first_entry), 1): #0,1,2 #len-1
    entry = first_entry[i]
    print("entry:",entry)
    print("df_output[key_list].iloc[i]:",df_output[key_list].iloc[i])
    df_output[key_list].iloc[i] = entry

  #df_map = pd.DataFrame(list(zip(df_tmp.values[0][0], df_tmp.values[0][1])),columns =[df_tmp.columns[0], df_tmp.columns[1]])
  #df_dropped = df.drop(columns = [key_list, val_list] )
  #df_output = pd.concat([df_dropped, df_map], axis=1)
  return df_output
```
